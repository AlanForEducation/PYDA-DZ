```python
#Теперь работаем только со столбцом type of lesion 1, его разобьем на 4:
# site of lesion
# type of lesion
# subtype
# specific code
# это проще всего сделать с помощью функции 
def split_type_of_lesinon(data):
    
    data = int(data)
    if data == 0:
        first = -999
        second = -999
        third = -999
        forth = -999
    else:
        
        data = str(data)
        if int(data[0:2]) == 11 and len(data) >= 5:
            first = int(data[0:2])
            second = int(data[2])
            third = int(data[3])
            forth = int(data[4:])
        elif len(data) >= 4:
            first = int(data[0])
            second = int(data[1])
            third = int(data[2])
            forth = int(data[3:])
        else:
            try:
                first = int(data[0])
            except:
                first = np.nan
            try:
                second = int(data[1])
            except:
                first = np.nan               
            try:
                third = int(data[2])
            except:
                third = np.nan
            try:
                forth = forth = int(data[3:])
            except:
                forth = np.nan
            
    res = [first, second, third, forth]   
    return res
```

```python
horses[['site of lesion', 'type of lesion', 'subtype', 'specific code']] = horses['type of lesion 1'].apply(split_type_of_lesinon).apply(pd.Series)
```

```python
horses['subtype'] = horses['subtype'].replace(3, np.nan)
horses['subtype'] = horses['subtype'].replace(0, np.nan)
horses['specific code'] = horses['specific code'].replace(0, np.nan)
```
