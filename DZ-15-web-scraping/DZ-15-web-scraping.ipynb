### Дополнительная часть (необязательная)

```python
import requests
import time
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
keywords = ['python', 'big data']
```

### Основная часть, поиск только по preview-информации

```python
url_ref = 'https://habr.com/ru/all/'
all_links = []
habr_news_p = pd.DataFrame()

for i in range(51):
    if i == 0:
        url = url_ref
    else:
        url = url_ref + 'page' + str(i)
        
    habr_res = requests.get(url)
    time.sleep(0.3)
    
    habr_soup = BeautifulSoup(habr_res.text, 'html.parser')
    news_blocks = habr_soup.find_all('li', class_='content-list__item content-list__item_post shortcuts_item')
    list_of_full_article_block = list(map(lambda x: x.find('article', class_='post'), news_blocks))
    for article in list_of_full_article_block:
        if article is not None:
            text = article.find('div', class_='post__text')
            text_lower = text.text.lower()
            if any([keyword in text_lower for keyword in keywords]):
                title = article.find('h2', class_='post__title').text
                user = article.find('span', class_='user-info__nickname').text
                link = article.find('h2', class_='post__title').find('a').get('href')
                date = article.find('span', class_='post__time').text

                row = {'date': date, 'user': user, 'title' : title, 'link' : link}
                habr_news_p = pd.concat([habr_news_p, pd.DataFrame([row])])

habr_news_p
```

### Дополнительное задание с поиском по всей статье

```python
# Получаем список из сслылок всех статей
# Так как на странице habr всегда можно перейти максиму на 50 страницу, чтобы не усложнять код использовал range

url_ref = 'https://habr.com/ru/all/'
all_links = []

for i in range(51):
    if i == 0:
        url = url_ref
    else:
        url = url_ref + 'page' + str(i)
        
    habr_res = requests.get(url)
    time.sleep(0.3)
    habr_soup = BeautifulSoup(habr_res.text, 'html.parser')
    news_blocks = habr_soup.find_all('li', class_='content-list__item content-list__item_post shortcuts_item')
    articles_intro = list(map(lambda x: x.find('h2', class_='post__title'), news_blocks))
    articles_intro = [x for x in articles_intro if x] #Удаляет пустые значения в списке
    all_links += list(map(lambda x: x.find('a').get('href'), articles_intro))
```

```python
habr_news = pd.DataFrame()
```

```python
#Создается data frame
#Делаю вывод i, чтобы контролировать процесс и понимать на каком этапе находится выполнение кода и понимать, что он не завис
#Habr блокировал после ~200 запросов(выдавал ошибку), если их делать с небольшим интервалом(0.3), 
# поэтому после каждых 100 запросов делаю большую паузу 

for i, link in enumerate(all_links):

    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    time.sleep(0.1)
    text = soup.find('div', class_='post__text')
    text_lower = text.text.lower()
    if any([keyword in text_lower for keyword in keywords]):

        date = pd.to_datetime(soup.find('span', class_='post__time').get('data-time_published'), dayfirst=True).date()
        user = soup.find('span', class_='user-info__nickname').text
        if 'post' in link:
            title = soup.find('span', class_='post__title-text').text

        elif 'company' in link:
            title = soup.find('h1', class_='post__title').text
        row = {'date' : date, 'user' : user, 'title' : title, 'text' : text.text}
        habr_news = pd.concat([habr_news, pd.DataFrame([row])])
    if i % 200 == 0:
        time.sleep(5.0)
    if i % 50 == 0:
        print(i)     
```

### Дополнительная часть (необязательная)
Написать скрипт, который будет получать 50 последних постов указанной группы во Вконтакте.
Документация к API VK: https://vk.com/dev/methods , вам поможет метод wall.getGROUP = 'netology'
TOKEN = УДАЛЯЙТЕ В ВЕРСИИ ДЛЯ ПРОВЕРКИ, НА GITHUB НЕ ВЫКЛАДЫВАТЬ

В итоге должен формироваться датафрейм со столбцами: <дата поста> - <текст поста>.

```python
import json
```

```python
emails = ['xxx@x.ru', 'yyy@y.com', 'prtmskjsl@gmail.com', 'ssss@gmail.com']
url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'

headers = {
    
    'Vaar-Version': '0',
    'Vaar-Header-App-Product': 'hackcheck-web-avast'
}


avast_leaks = pd.DataFrame()

for email in emails:
    payload = {'emailAddresses': [email]}
    avast = requests.post(url, data=json.dumps(payload), headers=headers)
    time.sleep(0.3)
    
    avast_list = avast.json()['breaches'].values()
    for leak in avast_list:
        date = leak['publishDate']
        source = leak['site']
        description = leak['description']

        row = {'email': email, 'date': date, 'source': source, 'description' : description}
        avast_leaks = pd.concat([avast_leaks, pd.DataFrame([row])])
        
avast_leaks
```
