## Вопрос 1
**Как критики относятся к спортивным играм?**

```python
sports_score = games[(games['Genre'] == 'Sports')&(games['Critic_Score'].notnull() == True)][['Critic_Score']]
sports_score.describe()
```

H0: критики высоко оценивают спортивные игры, ставя оценки больше 8  
H1: средняя оценка критиков меньше 8

Сравниваем значение по выборке с константой, поэтому нужен одновыборочный критерий.

Тест нужен односторонний, т.к. нас интересует только неравенство в меньшую сторону.

```python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(sports_score, 8)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, среднее меньше 8')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критики действительно ставят высокие оценки')
```

**Вывод:** Отвергаем нулевую гипотезу, результаты теста статистически значимы и средние оценики критиков оказались ниже 8

## Вопрос 2
**Критикам нравятся больше игры на PC или на PS4?**

```python
pc_or_ps = games[(games['Platform'].isin(['PC', 'PS4']))&(games['Critic_Score'].notnull() == True)][['Platform','Critic_Score']]
pc_or_ps
```

H0: Критики одинаково относятся к играм на pc и ps  
H1: У критиков есть предпочтения

Применяем двухсторонний двухвыборочный тест

```python
result = st.ttest_ind(pc_or_ps[pc_or_ps['Platform'] == 'PC']['Critic_Score'], pc_or_ps[pc_or_ps['Platform'] == 'PS4']['Critic_Score'], equal_var = False)
print(result)

if (result.pvalue < alpha):
    
    if(result.statistic < 0):
        print('Отвергаем нулевую гипотезу, критики предпочитают PS4')
        
    elif(result.statistic > 0):
        print('Отвергаем нулевую гипотезу, критики предпочитают PC')
else:
    print('Не отвергаем нулевую гипотезу, критики однаково относятся к играм на PS и PC')
```

**Вывод:** Значение p-value позволяет нам отклонить нулевую гипотезу, по значению statistic смотрим кому конкретно отдают предпочтения критики

## Вопрос 3
**Критикам больше нравятся стрелялки или стратегии?**

```python
shooter_strategy = games[(games['Genre'].isin(['Shooter', 'Strategy']))&(games['Critic_Score'].notnull() == True)][['Genre','Critic_Score']]
shooter_strategy
```

H0: Критики одинаково относятся стрелялкам и стратегиям  
H1: У критиков есть предпочтения

Применяем двухсторонний двухвыборочный тест

```python
result = (st.ttest_ind(shooter_strategy[shooter_strategy['Genre'] == 'Shooter']['Critic_Score'], 
                       shooter_strategy[shooter_strategy['Genre'] == 'Strategy']['Critic_Score'], 
                       equal_var = False))
print(result)

if (result.pvalue < alpha):
    
    if(result.statistic < 0):
        print('Отвергаем нулевую гипотезу, критики предпочитают Strategy')
        
    elif(result.statistic > 0):
        print('Отвергаем нулевую гипотезу, критики предпочитают Shooter')
else:
    print('Не отвергаем нулевую гипотезу, критики однаково относятся к стрелялкам и стратегиям')
```

**Вывод:** По statistic можно сказать, что есть перевес в пользу Стратегий, но он статистически не значимый, поэтому оставляем нулевую гипотезу, то есть критики однаково относятся к стрелялкам и стратегиям

# Задание 2

```python
import glob
import re
```

```python
spam = pd.read_csv('C:\\Users\\roman\\Learning\\PYDA13\\data\\spam.csv')
```

```python
# Приводим текст к нижнему регистру
spam['Message'] = spam['Message'].apply(lambda x: x.lower())

spam['Str_len_1'] = spam['Message'].str.len()
```

```python
# Удаляем мусорные символы
spam['Message'] = spam['Message'].apply(lambda x: re.sub('[\W_]+',' ', x))

spam['Str_len_2'] = spam['Message'].str.len()
```

```python
# проверяем отработала ли наша функция
spam.describe()
```

```python
spam.head()
```

Длинна стала меньше, следовательно, служебые символы удалились

```python
# Разбиваем сообщение на список из слов
spam['Message'] = spam['Message'].apply(lambda x: x.split())
```

```python
# Удаляем стоп слова
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
```

```python
# Функция, которая оставляет все, кроме стоп слов
def del_stop_words(message):
    new_message = []
    for word in message:
        if word not in stopwords_set:
            new_message += [word]
    return new_message
```

```python
# Применяем функцию
spam['Message'] = spam['Message'].apply(del_stop_words)
```

```python
# Приводим все слова к нормальной форме
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
# Функция для преведения слов к нормальной форме
def lemma(message):
    new_message = []
    for word in message:
        wordnet_lemmatizer.lemmatize(word)
        new_message += [word]
        
    return new_message
```

```python
# Применим функцию
spam['Message'] = spam['Message'].apply(lemma)
```

```python
# Удалим столбци, которые создавались для подсчета слов
del spam['Str_len_1']
del spam['Str_len_2']
```

```python
# Для преобразования сообщений в вектора, нужно сначало сделать каждое сообщение строкой
# TfidfVectorizer работает только со строками
spam['Message'] = spam['Message'].apply(lambda x: ' '.join(x))
```

```python
# Преобразуем сообщения в вектора

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(spam.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
```

```python
tfidf_matrix
```

```python
# Заменим категории цифрами для обучения нашей модели

spam['Category'] = spam['Category'].replace('ham', 0)
spam['Category'] = spam['Category'].replace('spam', 1)
```

```python
spam
```

```python
# Разбивам данные на тестовые и тренировочные и строим модель логистической регрессии 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Разбивам данные на тестовые и тренировочные
X = tfidf_matrix
y = spam['Category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# Строим модель логистической регрессии 
lr = LogisticRegression(random_state=42)  
lr.fit(X_train, y_train)
```

```python
# Прогоням модель на тестовых данных
y_pred = lr.predict(X_test)
```

```python
# Оцениваем результаты
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
lr.score(X_test, y_test)
```

**Описание результатов:** По confusion_matrix видим что модель ошиблась всего в 72 случаях(69+3), и с большой долей вероятности давала правильный результат

```python
# Строим датафрейм, который будет содержать все исходные тексты сообщений, классифицированные неправильно

# Разбиваем точно также наш исходный датафрейм
y = spam
y_train, y_test_df = train_test_split(y, test_size=0.30, random_state=42)
```

```python
# берем только сообщения, которые использовались для теста
result = y_test_df.copy()
```
